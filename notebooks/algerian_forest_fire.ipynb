## feature selection

```python
df=pd.read_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['month','day','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X
```

```python
y
```

```python
## train test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## feature selection based on corr 
X_train.corr()
```

```python
## checking for multicolinearity 
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True,)
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range (i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                col_name=corr_matrix.columns[i]
                col_corr.add(col_name)
    return col_corr
```

```python
## threshold but normally threshold is setup by domain expert or project manager 
corr_features=correlation(X_train,0.85)
```

```python
## drop these feature since due to correlation when corr is more than 0.85 

X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)

```

```python
X_train.shape,X_test.shape
```

## feature scaling or standardization 

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train.
```

```python
sns.boxplot(X_train_scaled)
```

```python
sns.boxplot(X_train)
```

# model Training 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score

linreg=LinearRegression(n_jobs=-1)
linreg
```

```python
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mae= mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
```

## apply lasso regression to fix with overfitting

```python
from sklearn.linear_model import Lasso

lasso=Lasso()

lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae= mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
```

### lasso cross validation 


```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV()
lassocv.fit(X_train_scaled,y_train)
```

```python
lassocv.alphas_
```

```python


lassocv.alpha_
```

```python
lassocv.mse_path_
```

```python
y_pred=lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(score)
```

# ridge regression 


```python
from sklearn.linear_model import Ridge

ridge=Ridge()

ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae= mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

ridgecv = RidgeCV(cv=5)  # Uses Leave-One-Out Cross-Validation by default
ridgecv.fit(X_train_scaled, y_train)

y_pred = ridgecv.predict(X_test_scaled)  #  Predict on test set

plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("RidgeCV: Actual vs Predicted")
plt.grid(True)
plt.show()

score = r2_score(y_test, y_pred)  #  Valid R score
print("R score on test set:", score)
```

## elastic net 

```python
from sklearn.linear_model import ElasticNet

elastic=ElasticNet()

elastic.fit(X_train_scaled,y_train)
y_pred=elastic.predict(X_test_scaled)
mae= mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import ElasticNetCV
elasticcv=ElasticNetCV(cv=5)

elasticcv.fit(X_train_scaled,y_train)
y_pred=elasticcv.predict(X_test_scaled)
mae= mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae)
print(score)
plt.scatter(y_test,y_pred)
```
